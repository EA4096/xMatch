```python
import sqlite3
import pandas as pd
import csv
import time
import numpy as np
```

База данных на данном этапе представлена двумя таблицами ps и sdss, содержащими в себе данные каталогов PS1 и SDSS DR12  о поле квадрата стороной $6^\circ$ с центром в $(243.9021249, 14.41539495)$:  ps1.csv, sdss12.csv

http://tapvizier.u-strasbg.fr/adql/ 

```python
conn=sqlite3.connect('catalogs.db') 
cur=conn.cursor()

# cur.execute('SELECT * FROM sqlite_master').fetchall()
```

```python
%%time 

ps_pand=pd.read_csv("ps1.csv", sep=',', low_memory=False)
ps_pand.astype(float)
ps_pand['ps_objID'].astype(int, copy = False)
ps_df = ps_pand[['ps_objID', 'ps_RA', 'ps_e_RA', 'ps_DE', 'ps_e_DE','ps_gmag', 'ps_e_gmag','ps_rmag', 'ps_e_rmag','ps_imag', 'ps_e_imag','ps_zmag', 'ps_e_zmag','ps_ymag', 'ps_e_ymag']]
ps_df
```

```python
%%time

sdss_pand=pd.read_csv("sdss12.csv", sep=',', low_memory=False)
sdss_pand.astype(float)
sdss_pand['sdss_objID'].astype(int, copy = False)

sdss_df = sdss_pand[['sdss_objID', 'sdss_RA', 'sdss_e_RA', 'sdss_DE', 'sdss_e_DE', 'sdss_umag', 'sdss_e_umag', 'sdss_gmag', 'sdss_e_gmag','sdss_rmag', 'sdss_e_rmag','sdss_imag', 'sdss_e_imag','sdss_zmag', 'sdss_e_zmag']]
sdss_df
```

```python
ps_df.to_sql(name='ps', con=conn, if_exists='replace', index=False)
```

```python
sdss_df.to_sql(name='sdss', con=conn, if_exists='replace', index=False)
```

```python
conn.commit()
```

```python
conn.close()
```
